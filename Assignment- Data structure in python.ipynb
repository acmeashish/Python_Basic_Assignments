```python
#1. Write code to reverse a string
a="hello world"
print(a[::-1])

#2. Write a code to count the number of vowels in a string
def count_vowels(string):
    vowels = "aeiouAEIOU"
    count = 0
    for char in string:
        if char in vowels:
            count += 1 
    return count
# Example usage
input_string = "ram has gone to the market for bring brinjal"
vowel_count = count_vowels(input_string)
print("Number of vowels:", vowel_count)


#3. Write a code to check if a given string is a palindrome or not.
def is_palindrome(string):
    # Remove spaces and convert to lowercase for uniformity
    clean_string = string.replace(" ", "").lower()
    
    # Check if the string is equal to its reverse
    return clean_string == clean_string[::-1]

# Example usage
input_string = "Never odd or even"
if is_palindrome(input_string):
    print(f'"{input_string}" is a palindrome.')
else:
    print(f'"{input_string}" is not a palindrome.')


#4.Write a code to check  if two given strings are anagrams of each other.
def is_anagram(str1,str2):
    # Remove spaces and convert to lowercase for uniformity
    str1=str1.replace(" ", "").lower()
    str2=str2.replace(" ", "").lower()
    
    # Sort the characters of both strings and compare
    return sorted(str1) == sorted(str2)

string1 = "Listen"
string2 = "Silent"

if is_anagram(string1, string2):
    print(f'"{string1}" and "{string2}" are anagrams.')
else:
    print(f'"{string1}" and "{string2}" are not anagrams.')

string1 = "Listen"
string2 = "ashish"

if is_anagram(string1, string2):
    print(f'"{string1}" and "{string2}" are anagrams.')
else:
    print(f'"{string1}" and "{string2}" are not anagrams.')

#5.Write a code to find all occurences of a given substring within another string.
def find_all_occurrences(string, substring):
    start = 0
    indices = []
    while True:
        # Use find() to locate the next occurrence of the substring
        start = string.find(substring, start)
        # If find() returns -1, no more occurrences are found
        if start == -1:
            break
        # Store the starting index of the found occurrence
        indices.append(start)
        # Move the start index forward to continue searching
        start += 1
    return indices
# Example usage:
main_string = "abracadabra"
sub_string = "abra"
result = find_all_occurrences(main_string, sub_string)
print(result)  # Output: [0, 7]

#6. Write a code to perform basic string compression using the counts of repreated characters.

def compress_string(s):
    # Initialize variables
    compressed = []
    count = 1

    # Loop through the string
    for i in range(1, len(s)):
        # If current character matches the previous one, increment count
        if s[i] == s[i - 1]:
            count += 1
        else:
            # Append the character and count to the compressed list
            compressed.append(s[i - 1] + str(count))
            count = 1

    # Append the last character and its count
    if s:
        compressed.append(s[-1] + str(count))

    # Convert list to string and return the shorter of the original and compressed
    compressed_string = ''.join(compressed)
    return compressed_string if len(compressed_string) < len(s) else s

# Test the function
input_string = "aabcccccaaa"
result = compress_string(input_string)
print(f"Original: {input_string}")
print(f"Compressed: {result}")

# 7. Write a code to determine if a string has all unique characters
def has_unique_characters(s):
    # Use a set to track seen characters
    seen = set()
    
    for char in s:
        if char in seen:
            return False  # Duplicate found
        seen.add(char)
    
    return True  # No duplicates found

# Test the function
input_string = "abcdef"
result = has_unique_characters(input_string)
print(f"String: {input_string}")
print(f"Has all unique characters: {result}")


#8.write a code to convert a given string to uppercase or lowercase.
a="pwskills"
print(a.upper())
print(a.lower())

#9. write a code to count the number of words in strings.
state="delhi"
print(len(state))

#10. Write a code to concatenate two string without using + operator.
def concatenate_strings_fstring(str1, str2):
    return f"{str1}{str2}"

result = concatenate_strings_fstring("Hello ", "World")
print(result)

#11. Write a code to remove all occurrences of a specific element from a list

def remove_element(lst, element):
    # Use a list comprehension to filter out the specified element
    return [x for x in lst if x != element]

# Test the function
input_list = [1, 2, 3, 4, 2, 5, 2]
element_to_remove = 2

result = remove_element(input_list, element_to_remove)
print(f"Original List: {input_list}")
print(f"Element to Remove: {element_to_remove}")
print(f"Updated List: {result}")

#12. Implement a code to find the second largest number in a given list of integers
def find_second_largest(nums):
    if len(nums) < 2:
        return None  # Not enough elements to determine second largest
    
    # Initialize variables to track the largest and second largest
    largest = second_largest = float('-inf')
    
    for num in nums:
        if num > largest:
            # Update both largest and second largest
            second_largest = largest
            largest = num
        elif num > second_largest and num != largest:
            # Update second largest only if the number is not equal to largest
            second_largest = num
    
    return second_largest if second_largest != float('-inf') else None

# Test the function
input_list = [10, 20, 4, 45, 99, 45]
result = find_second_largest(input_list)
print(f"List: {input_list}")
print(f"Second Largest: {result}")

#13.Create a code to count the occurrences of each element in a list and return a dictionary with elements as keys and their counts as values
def count_occurrences(lst):
    # Use a dictionary to store counts
    counts = {}
    
    for element in lst:
        if element in counts:
            counts[element] += 1
        else:
            counts[element] = 1
    
    return counts

# Test the function
input_list = [1, 2, 3, 2, 1, 5, 1, 2, 4, 5, 5]
result = count_occurrences(input_list)
print(f"List: {input_list}")
print(f"Occurrences: {result}")

#14.Write a code to reverse a list in-place without using any built-in reverse functions
def reverse_list(lst):
    # Use two-pointer technique
    left = 0
    right = len(lst) - 1
    
    while left < right:
        # Swap elements at the two pointers
        lst[left], lst[right] = lst[right], lst[left]
        # Move the pointers closer to the center
        left += 1
        right -= 1

# Test the function
input_list = [1, 2, 3, 4, 5]
print(f"Original List: {input_list}")
reverse_list(input_list)
print(f"Reversed List: {input_list}")

#15.Implement a code to find and remove duplicates from a list while preserving the original order of elements
def remove_duplicates(lst):
    seen = set()  # To track already encountered elements
    result = []  # To store the final list with duplicates removed
    
    for item in lst:
        if item not in seen:
            result.append(item)
            seen.add(item)
    
    return result

# Test the function
input_list = [1, 2, 3, 2, 1, 5, 1, 2, 4, 5, 5]
result = remove_duplicates(input_list)
print(f"Original List: {input_list}")
print(f"List after removing duplicates: {result}")

#16.Create a code to check if a given list is sorted (either in ascending or descending order) or not
def is_sorted(lst):
    # Check if the list is empty or has a single element
    if len(lst) <= 1:
        return True

    # Check for ascending order
    ascending = all(lst[i] <= lst[i + 1] for i in range(len(lst) - 1))
    # Check for descending order
    descending = all(lst[i] >= lst[i + 1] for i in range(len(lst) - 1))

    if ascending:
        return "Ascending"
    elif descending:
        return "Descending"
    else:
        return "Not Sorted"

# Test the function
input_list = [1, 2, 3, 4, 5]
result = is_sorted(input_list)
print(f"List: {input_list}")
print(f"Sorted in: {result}")

# Test with descending order
input_list2 = [5, 4, 3, 2, 1]
result2 = is_sorted(input_list2)
print(f"List: {input_list2}")
print(f"Sorted in: {result2}")

# Test with unsorted list
input_list3 = [1, 3, 2, 5, 4]
result3 = is_sorted(input_list3)
print(f"List: {input_list3}")
print(f"Sorted in: {result3}")

#17.Write a code to merge two sorted lists into a single sorted list
def merge_sorted_lists(list1, list2):
    merged_list = []
    i = j = 0
    
    # Merge the lists until one of them is exhausted
    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            merged_list.append(list1[i])
            i += 1
        else:
            merged_list.append(list2[j])
            j += 1
    
    # If there are remaining elements in list1, add them
    while i < len(list1):
        merged_list.append(list1[i])
        i += 1
    
    # If there are remaining elements in list2, add them
    while j < len(list2):
        merged_list.append(list2[j])
        j += 1
    
    return merged_list

# Test the function
list1 = [1, 3, 5, 7]
list2 = [2, 4, 6, 8]
result = merge_sorted_lists(list1, list2)
print(f"List 1: {list1}")
print(f"List 2: {list2}")
print(f"Merged List: {result}")

#18.Implement a code to find the intersection of two given lists
def intersection_of_lists(list1, list2):
    # Use set to find the intersection (removes duplicates and ensures O(1) lookup)
    set1 = set(list1)
    set2 = set(list2)
    
    # Find common elements between set1 and set2
    intersection = set1 & set2
    
    return list(intersection)

# Test the function
list1 = [1, 2, 3, 4, 5]
list2 = [3, 4, 5, 6, 7]
result = intersection_of_lists(list1, list2)
print(f"List 1: {list1}")
print(f"List 2: {list2}")
print(f"Intersection: {result}")

#19.Create a code to find the union of two lists without duplicates
def union_of_lists(list1, list2):
    # Use set to combine the lists and remove duplicates
    return list(set(list1) | set(list2))

# Test the function
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]
result = union_of_lists(list1, list2)
print(f"List 1: {list1}")
print(f"List 2: {list2}")
print(f"Union: {result}")

#20.Write a code to shuffle a given list randomly without using any built-in shuffle functions
import random

def shuffle_list(lst):
    # Get the length of the list
    n = len(lst)
    
    # Shuffle using random swapping technique
    for i in range(n):
        # Select a random index from i to the end of the list
        j = random.randint(i, n - 1)
        
        # Swap the elements at index i and index j
        lst[i], lst[j] = lst[j], lst[i]
    
    return lst

# Test the function
input_list = [1, 2, 3, 4, 5]
result = shuffle_list(input_list)
print(f"Original List: [1, 2, 3, 4, 5]")
print(f"Shuffled List: {result}")

#21.Write a code that takes two tuples as input and returns a new tuple containing elements that are common to both input tuples
def common_elements(tuple1, tuple2):
    # Convert tuples to sets and find the intersection
    common_set = set(tuple1) & set(tuple2)
    
    # Convert the result back to a tuple
    return tuple(common_set)

# Test the function
tuple1 = (1, 2, 3, 4, 5)
tuple2 = (3, 4, 5, 6, 7)
result = common_elements(tuple1, tuple2)
print(f"Tuple 1: {tuple1}")
print(f"Tuple 2: {tuple2}")
print(f"Common Elements: {result}")

#22.Create a code that prompts the user to enter two sets of integers separated by commas. Then, print the intersection of these two sets
def get_set_from_input(prompt):
    # Get input from user, split by commas, and convert to a set of integers
    user_input = input(prompt)
    # Convert the input string to a list of integers and then to a set
    return set(map(int, user_input.split(',')))

def intersection_of_sets(set1, set2):
    # Find and return the intersection of the two sets
    return set1 & set2

# Prompt user to enter two sets
set1 = get_set_from_input("Enter the first set of integers (separated by commas): ")
set2 = get_set_from_input("Enter the second set of integers (separated by commas): ")

# Find and print the intersection
intersection = intersection_of_sets(set1, set2)
print(f"The intersection of the two sets is: {intersection}")

#23.Write a code to concatenate two tuples. The function should take two tuples as input and return a new tuple containing elements from both input tuples.
def concatenate_tuples(tuple1, tuple2):
    # Concatenate the two tuples using the '+' operator
    return tuple1 + tuple2

# Test the function
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)
result = concatenate_tuples(tuple1, tuple2)
print(f"Tuple 1: {tuple1}")
print(f"Tuple 2: {tuple2}")
print(f"Concatenated Tuple: {result}")

#24.Develop a code that prompts the user to input two sets of strings. Then, print the elements that are present in the first set but not in the second set&
def get_set_from_input(prompt):
    # Get input from user, split by commas, and convert to a set of strings
    user_input = input(prompt)
    # Convert the input string to a list and then to a set
    return set(user_input.split(','))

def difference_of_sets(set1, set2):
    # Find the elements present in set1 but not in set2
    return set1 - set2

# Prompt user to enter two sets of strings
set1 = get_set_from_input("Enter the first set of strings (separated by commas): ")
set2 = get_set_from_input("Enter the second set of strings (separated by commas): ")

# Find and print the difference
difference = difference_of_sets(set1, set2)
print(f"The elements in the first set but not in the second set are: {difference}")

#25.Create a code that takes a tuple and two integers as input. The function should return a new tuple containing elements from the original tuple within the specified range of indices&
def slice_tuple(original_tuple, start_index, end_index):
    # Slice the tuple from start_index to end_index (exclusive)
    return original_tuple[start_index:end_index]

# Test the function
original_tuple = (10, 20, 30, 40, 50, 60)
start_index = int(input("Enter the start index: "))
end_index = int(input("Enter the end index: "))

result = slice_tuple(original_tuple, start_index, end_index)
print(f"Original Tuple: {original_tuple}")
print(f"Sliced Tuple: {result}")

#26.Write a code that prompts the user to input two sets of characters. Then, print the union of these two sets&
def get_set_from_input(prompt):
    # Get input from user, split by commas, and convert to a set of characters
    user_input = input(prompt)
    # Convert the input string to a list and then to a set
    return set(user_input.split(','))

def union_of_sets(set1, set2):
    # Find and return the union of the two sets
    return set1 | set2

# Prompt user to enter two sets of characters
set1 = get_set_from_input("Enter the first set of characters (separated by commas): ")
set2 = get_set_from_input("Enter the second set of characters (separated by commas): ")

# Find and print the union
union = union_of_sets(set1, set2)
print(f"The union of the two sets is: {union}")

#27.Develop a code that takes a tuple of integers as input. The function should return the maximum and minimum values from the tuple using tuple unpacking&
def find_max_min(input_tuple):
    # Use tuple unpacking to return max and min values from the tuple
    return max(input_tuple), min(input_tuple)

# Test the function
input_tuple = tuple(map(int, input("Enter a tuple of integers (separated by commas): ").split(',')))

max_value, min_value = find_max_min(input_tuple)

print(f"Tuple: {input_tuple}")
print(f"Maximum Value: {max_value}")
print(f"Minimum Value: {min_value}")

#28.Create a code that defines two sets of integers. Then, print the union, intersection, and difference of these two sets&
def set_operations(set1, set2):
    # Calculate union, intersection, and difference
    union = set1 | set2
    intersection = set1 & set2
    difference = set1 - set2
    
    return union, intersection, difference

# Define two sets of integers
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

# Perform the set operations
union, intersection, difference = set_operations(set1, set2)

# Print the results
print(f"Set 1: {set1}")
print(f"Set 2: {set2}")
print(f"Union: {union}")
print(f"Intersection: {intersection}")
print(f"Difference (Set1 - Set2): {difference}")

#29.Write a code that takes a tuple and an element as input. The function should return the count of occurrences of the given element in the tuple&
def count_occurrences(input_tuple, element):
    # Use the count() method to find the occurrences of the element in the tuple
    return input_tuple.count(element)

# Test the function
input_tuple = tuple(map(int, input("Enter a tuple of integers (separated by commas): ").split(',')))
element = int(input("Enter the element to count: "))

count = count_occurrences(input_tuple, element)

print(f"Tuple: {input_tuple}")
print(f"Element: {element}")
print(f"Count of occurrences: {count}")

#30.Develop a code that prompts the user to input two sets of strings. Then, print the symmetric difference of these two sets
def symmetric_difference(set1, set2):
    # Calculate the symmetric difference of the two sets
    return set1 ^ set2

# Prompt the user to input two sets of strings
set1 = set(input("Enter the first set of strings (separated by spaces): ").split())
set2 = set(input("Enter the second set of strings (separated by spaces): ").split())

# Perform the symmetric difference operation
sym_diff = symmetric_difference(set1, set2)

# Print the result
print(f"Set 1: {set1}")
print(f"Set 2: {set2}")
print(f"Symmetric Difference: {sym_diff}")

#31.Write a code that takes a list of words as input and returns a dictionary where the keys are unique words and the values are the frequencies of those words in the input list&
def word_frequencies(word_list):
    # Create an empty dictionary to store word frequencies
    frequency_dict = {}
    
    # Iterate over the words in the list
    for word in word_list:
        if word in frequency_dict:
            frequency_dict[word] += 1
        else:
            frequency_dict[word] = 1
    
    return frequency_dict

# Prompt the user to input a list of words (separated by spaces)
input_list = input("Enter a list of words (separated by spaces): ").split()

# Get the word frequencies
frequencies = word_frequencies(input_list)

# Print the result
print(f"Word Frequencies: {frequencies}")

#32.Write a code that takes two dictionaries as input and merges them into a single dictionary. If there are common keys, the values should be added together
def merge_dictionaries(dict1, dict2):
    # Create a new dictionary to store the merged result
    merged_dict = dict1.copy()  # Start with a copy of the first dictionary
    
    # Iterate over the second dictionary and merge
    for key, value in dict2.items():
        if key in merged_dict:
            merged_dict[key] += value  # Add the values for common keys
        else:
            merged_dict[key] = value  # Add new key-value pairs

    return merged_dict

# Input: Prompt the user to input two dictionaries
dict1 = eval(input("Enter the first dictionary: "))
dict2 = eval(input("Enter the second dictionary: "))

# Merge the dictionaries
merged_result = merge_dictionaries(dict1, dict2)

# Print the merged dictionary
print(f"Merged Dictionary: {merged_result}")

#33.Write a code to access a value in a nested dictionary. The function should take the dictionary and a list of keys as input, and return the corresponding value. If any of the keys do not exist in the dictionary, the function should return None
def get_nested_value(nested_dict, keys):
    # Iterate over the list of keys and navigate through the nested dictionary
    for key in keys:
        # Check if the current key exists in the dictionary
        if key in nested_dict:
            nested_dict = nested_dict[key]  # Move to the next level in the nested dictionary
        else:
            return None  # Return None if any key does not exist
    return nested_dict  # Return the value if all keys are found

# Input: A nested dictionary and a list of keys
nested_dict = eval(input("Enter the nested dictionary: "))
keys = input("Enter a list of keys (separated by commas): ").split(',')

# Convert keys to remove extra spaces and handle strings properly
keys = [key.strip() for key in keys]

# Get the corresponding value using the function
result = get_nested_value(nested_dict, keys)

# Print the result
print(f"Result: {result}")

#34.Write a code that takes a dictionary as input and returns a sorted version of it based on the values. You can choose whether to sort in ascending or descending order
def sort_dict_by_values(input_dict, descending=False):
    # Sort the dictionary by its values and return a new sorted dictionary
    sorted_dict = dict(sorted(input_dict.items(), key=lambda item: item[1], reverse=descending))
    return sorted_dict

# Input: A dictionary
input_dict = eval(input("Enter a dictionary: "))

# Input: Ask user whether to sort in ascending or descending order
order = input("Do you want to sort in descending order? (yes/no): ").strip().lower()

# Convert user's choice to boolean (True for descending, False for ascending)
descending = order == 'yes'

# Get the sorted dictionary
sorted_dict = sort_dict_by_values(input_dict, descending)

# Print the sorted dictionary
print(f"Sorted Dictionary: {sorted_dict}")

#35.Write a code that inverts a dictionary, swapping keys and values. Ensure that the inverted dictionary correctly handles cases where multiple keys have the same value by storing the keys as a list in the inverted dictionary.
def invert_dict(input_dict):
    inverted_dict = {}

    for key, value in input_dict.items():
        # If the value is already a key in the inverted dictionary, append the key to the list
        if value in inverted_dict:
            inverted_dict[value].append(key)
        else:
            inverted_dict[value] = [key]

    return inverted_dict

# Input: A dictionary
input_dict = eval(input("Enter a dictionary: "))

# Invert the dictionary
inverted_dict = invert_dict(input_dict)

# Print the inverted dictionary
print(f"Inverted Dictionary: {inverted_dict}")

```
